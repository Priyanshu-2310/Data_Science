## Copy in set

## why you need copy method

```python
a = {12, 7, 8, 60, 21}
b = a
b.add(10)
print(a)
print(b)

# it change original set as well
```

```python
a = {12, 3, 4, 5, 6}
b = a.copy()

b.add(21)
print(a)
print(b)
```

## 1. if Condition

```python
"""take two value from user and compare them to find which one is greater..."""
num1 = int(input("Enter The first value = "))
num2 = int(input("Enter the second value = "))
if(num1>num2):
    print("Number one is greater", num1)
if(num2>num1):
    print("Number two is greater = ", num2)
```

```python
"""Enter the value from user and check that number is even or odd"""
num1 = int(input("Enter the number = "))
if(num1%2==0):
    print("Number is even")
if(num1%2!=0):
    print("Number is odd")
```

```python
"""take number from user and check that number is divisible by 13 or not.."""
num1 = int(input("enter the number = "))
if(num1%13==0):
    print("Number is divisible by 13")
if num1%13!=0:
    print("Not divisible by 13")
```

## 2 . if else

```python
"""take a value from user if the value is 20 print magic and otherwise print no magic"""
num = int(input("Enter the number = "))
if(num == 20):
    print("magic")
else:
    print("No Magic")
```

```python
"""Enter the age from user if age is greater than 60 than ticket is freee other wise you will have to pay"""
age = int(input("Enter the age = "))
if(age ==60):
    print("Free")
else:
    print("Pay for it")
```

```python
"""check user is eligible for vote or not"""
age = int(input("enter the age = "))
if(age >= 18):
    print("eligible")
else:
    print("not eligible")
```

## 3.  if elif else

```python
"""take two value from user if first is greater print greater other wise lesser otherwise equal"""
num1 = int(input("Enter the first value = "))
num2 = int(input("Enter the second value = "))
if(num1 > num2):
    print("greater")
elif(num2 > num1)
    print("lesser")
else:
    print("equal")
```

```python
"""take the value from user and check either value is divisible by 2 or 3 or neigther divisible by 2 or 3"""
num1 = int(input("Enter the value = "))
if(num1 % 2 == 0):
    print("value is divisible by 2")
elif(num1 % 3 == 0):
    print("value is divisible by 3")
else:
    print("value is not divisible by 2 and 3")
```

```python
"""Enter the marks of student if marks is greater than 85 so A if marks is greater than 90 so A+ and if marks is greater than 70 so B and if marks is greater than 55 so c if marks is greater than 45 so d if marks is greater than 35 so f"""
marks = int(input("Enter the marks = "))
if(marks >= 90):
    print("A+")
elif(marks >= 80 and marks < 90):
    print("A")
elif(marks >=70 and marks < 80):
    print("B")
elif(marks >=55 and marks<70):
    print("C")
elif(marks >= 45 and marks < 55):
    print("E")
else:
    print("F")
```

## Nested if 

```python
"""Enter the value from user and check it is greater than 10 or 20 and not""" 

num = int(input("Enter the number = "))

if(num > 10):
    print("Inside the if block")
    if(num>20):
        print("Inside the nested if")
    else:
        print("Inside the nested else")
else:
    print("Inside the else")
```

## Break Statement

```python
num = 9
i = 1
while True:
    if (i % num == 0):
        print("Inside if")
        break
    else :
        print("Inside else")
        i += 1

print('done')
```

## For loops

```python
"""Write a program and to print a square of a number till 10"""
L = []
for x in range (1, 11):
    print(x)
    L.append(x**2)

print(L)
```

```python
L = []
for x in range (0, 10, 2):
    print(x)
    L.append(x**2)

print(L)
```

```python
"""Write a program to separate negative and positive value from list"""
numbers = [7, -3, 12, -8, 0, 5, -1, 9, -6, 2]
pos = []
neg = []

for x in range (0, len(numbers)):
    if(numbers[x] >= 0):
        pos.append(numbers[x])
    else:
        neg.append(numbers[x])

print(pos)
print(neg)
```

```python
"""Write a program to separate even and odd number from list """
numbers = [7, 12, 3, 8, 15, 2, 9, 10, 5, 6, 13, 4, 11, 14, 1]
evenlist = []
oddlist = []

for x in range (0, len(numbers)):
    if(numbers[x] % 2 == 0):
        evenlist.append(numbers[x])
    else:
        oddlist.append(numbers[x])

print(evenlist)
print(oddlist)
```

## Nested loop

```python
"""Ask a user to enter a string and filter count vowel in the string"""
L = []
str = input("Enter the string = ")

for x in range(len(str)):
    L.append(str[x])

print(L)

count = 0
for i in L:
    if(i == 'a'):
        count = count + 1;

print("a = ", count)

count1 = 0
for i in L:
    if(i == 'e'):
        count1 = count1 + 1;

print("e = ", count1)

count2 = 0
for i in L:
    if(i == 'i'):
        count2 = count2 + 1

print("i = ", count2)

count3 = 0
for i in L:
    if(i == 'o'):
        count3 = count3 + 1

print("O = ", count3)

count4 = 0
for i in L:
    if(i == 'u'):
        count4 = count4 + 1

print("u = ", count4)
    
```

## Min And Max Function

```python
list2 = [1, 2, 2, 1, 5, 6, 3, 343, 22]
print(min(list2))
print(max(list2))
```

```python
"""Q1. Ask the user to enter three values and find the maximum and minimum among them"""

val1 = input("enter value = ")
val2 = input("enter value = ")
val3 = input("enter value = ")
val4 = input("enter value = ")
val5 = input("enter value = ")

maximum = max(val1, val2, val3, val4, val5)
minimum = min(val1, val2, val3, val4, val5)

print(maximum, minimum)
```

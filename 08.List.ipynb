## Slicing 

```python
a = [12, 13, 13, 13, 12 , "hello", "hii", "kaise ho"]
print(a[0: 5])
print(a[5: ])
```

```python
a = [12, 13, 13, 13, 12 , "hello", "hii", "kaise ho"]
b = a[0: 5]
c = a[5: ]
print(b)
print(c)
```

```python
"""Divide a list into two and reverse it """
a = [12, 13, 14, 14, "hello", "hii", "helloooo"]
b = a[0: 4]
c = a[5: ]
print(b)
print(c)
print(b[: : -1])
print(c[: : -1])
```

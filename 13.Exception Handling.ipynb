## Exception Handling

```python
a = int(input("Enter your first number = "))
b = int(input("Enter your second number = "))

print(a/b)

print(a + b)
```

```python
a = int(input("Enter your first number = "))
b = int(input("Enter your second number = "))

try: 
    print(a/b)

except Exception as err:
    print(f"Error occured :  {err}")

```

## Else Keyword

```python
a = int(input("Enter your first number = "))
b = int(input("Enter your second number = "))

try: 
    print(a/b)

except Exception as err:
    print(f"Error occured :  {err}")

else:
    print("No Error Occured")
```

## Finally keyword

```python
a = int(input("Enter your first number = "))
b = int(input("Enter your second number = "))

try: 
    print(a/b)

except Exception as err:
    print(f"Error occured :  {err}")

else:
    print("No Error Occured")

finally: 
    print("Always run")
```

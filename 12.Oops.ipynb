```python
class bank_account:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance = self.__balance + amount

    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance = self.__balance - amount
        else:
            return False

    def get_balance(self):
        return self.__balance
```

```python
priyanshu = bank_account(1000)
```

```python
priyanshu.get_balance()
```

```python
priyanshu.deposit(4000)
```

```python
priyanshu.get_balance()
```

## Multilevel Inheritance

```python
class class1():
    def test(self):
        print("this is the method of class 1")
```

```python
class class2(class1):
    def learn(self):
        print("this is method of class 2")
```

```python
class class3(class2):
    pass
```

```python
cls3 = class3()
```

```python
cls3.learn()
```

```python
cls3.test()
```

## Multiple inheritance

```python
class class1():
    def fun1(self):
        print("this is class 1")
```

```python
class class2():
    def fun2(self):
        print("this is class2")
```

```python
class class3(class1, class2):
    pass
```

```python
class3_obj = class3()
```

```python
class3_obj.fun2()
```

## Abstraction

```python
import abc

class ihh():
    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def homework_details(self):
        pass

    @abc.abstractmethod
    def exam_details(self):
        pass
```

```python
class demo(ihh):
    def student_details(self):
        print("student detials are avalible here")

    def homework_details(self):
        print(4 + 5)
```

```python
class test(ihh):
    def student_details(self):
        print("student is enrolled")

    def exam_details(self):
        print("your exam is on next week")
```

```python
demo_obj = demo()
```

```python
demo_obj.student_details()
```

## Using loop in tuple

```python
"""Seperate positive and negative intergers from the tuple"""

a = (14, 6, 17, -83, -45, -32, 12, 43, -54)
pos = []
neg = []

for x in a:
    if(x>0):
        pos.append(x)
    else:
        neg.append(x)

post = tuple(pos)
negt = tuple(neg)

print(post)
print(negt)
```

```python
"""Multiply adjacent element in tuple"""

a = (2, 5, 3, 6, 4, 8, 1)
mul = []
for x in range(len(a)-1):
    multiply = a[x] * a[x+1]
    mul.append(multiply)

multuple = tuple(mul)
print(multuple)
    
    
```

```python
"""Original list of tuple  [(1, 2), (2, 3), (3, 4)]
convert it into [13, 5, 7]
"""

listtuple = [(1, 2), (2, 3), (3, 4)]
list1 = []
for x in listtuple:
    list1.append(sum(x))

print(list1)
```

## Variable number of input Arguments

```python
def argfunc(*args):
    sum = 0;
    for x in range(len(args)):
        sum = sum + args[x]
    print(sum)
```

```python
argfunc(12, 3, 4, 5,)
```

## Keyword Variable number of input arguments

```python
def kwargfunc(**kwargs):
    for x in kwargs:
        print(x, kwargs[x])
```

## Generator Function

```python
def genfun():
    yield 1
    yield 2
    yield 3
```

```python
print(genfun())
```

```python
result = genfun()
print(next(result))
print(next(result))
print(next(result))
```

```python
def febo(n):
    a = 0
    b = 1
    for x in range(n):
        yield a
        a, b = b, a + b


```

```python
for x in febo(5):
    print(x)
```
